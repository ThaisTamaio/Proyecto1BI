```python
# Reemplazar los valores 'positivo' por 1 y 'negativo' por 0 en la columna 'sentimiento'
movies_df['sentimiento'] = movies_df['sentimiento'].replace({'positivo': 1, 'negativo': 0})
```

Se eliminan smbolos de puntuacin o comillas para evitar que los caracteres de puntuacin interfieran en la interpretacin de las palabras:

```python
def eliminar_puntuacion(review):
    review = review.translate(str.maketrans(string.punctuation + '', ' ' * len(string.punctuation + '')))
    return review
```

```python
movies_df['review_es'] = movies_df['review_es'].apply(eliminar_puntuacion)
```

La funcin limpiar_review toma un string de texto (una revisin) como entrada y realiza una serie de operaciones para limpiar el texto. El objetivo detrs de esto es preparar el texto para el anlisis de sentimiento o cualquier otro tipo de anlisis de texto. En particular, se eliminan caracteres que no sean ASCII, se pasan todos los caracteres a minsculas, se eliminan puntuaciones y se transforman los nmeros en su representacin textual en espaol. Adems, se eliminan stopwords y palabras que no sean en espaol. El resultado final es una versin limpia y preprocesada del texto que puede ser utilizada para anlisis posteriores.

```python
def limpiar_review(review):
    # Eliminar caracteres que no sean ASCII
    review = unicodedata.normalize('NFKD', review).encode('ascii', 'ignore').decode('utf-8')
    
    # Pasar todos los caracteres a minsculas
    review = review.lower()
    
    # Remover puntuacin
    review = review.translate(str.maketrans('', '', string.punctuation))
    
    # Transformar valores numricos dentro de las palabras a su representacin textual en espaol
    review = re.sub(r'\b\d+\b', lambda x: num2words(int(x.group(0)), lang='es'), review)
    
    # Eliminar stopwords y palabras que no sean en espaol
    review = [w for w in review.split() if w not in spanish_stopwords]
    
    return " ".join(review) # Unir los tokens en un solo string separados por espacios
```

El uso de la librera "Parallel" y el argumento "n_jobs=-1" permite realizar esta tarea de forma paralela, acelerando significativamente el proceso en computadoras con varios ncleos.

```python
movies_df['review_es'] = Parallel(n_jobs=-1)(delayed(limpiar_review)(review) for review in movies_df['review_es'])
```

# 3. Random Forest

Random Forest es un algoritmo de aprendizaje supervisado que se utiliza tanto para la clasificacin como para la regresin. Este algoritmo se basa en la combinacin de mltiples rboles de decisin, donde cada rbol se entrena con una submuestra aleatoria del conjunto de datos de entrenamiento. Durante la prediccin, el algoritmo promedia las predicciones de todos los rboles de decisin para obtener la salida final.

Random Forest se puede utilizar en la clasificacin de reseas en positivas y negativas porque es capaz de manejar mltiples caractersticas de texto (como BoW, TF-IDF y HashingVectorizer) al mismo tiempo, lo que puede mejorar el rendimiento en la clasificacin. Adems, al combinar mltiples rboles de decisin, Random Forest reduce el riesgo de sobreajuste y mejora la generalizacin del modelo a nuevos datos. En resumen, Random Forest puede ser til para entrenar modelos de clasificacin de reseas porque puede manejar mltiples caractersticas de texto, reduce el riesgo de sobreajuste y mejora la generalizacin del modelo a nuevos datos.

## 3.2 Modelo BoW (Juan Camilo Bonet)

### 3.2.1 Vectorizando reviewes segn BoW

A continuacin, se crea una representacin numrica de los textos en X_train mediante el enfoque de bolsa de palabras, con el propsito de utilizarla para entrenar un modelo de aprendizaje automtico para anlisis de sentimientos.

La representacin en forma de bolsa de palabras (BoW) cuenta la frecuencia de cada palabra en cada texto y luego genera una matriz de caractersticas que se puede utilizar como entrada para un modelo de aprendizaje automtico. El objetivo es convertir los textos en datos numricos para que puedan ser procesados por el modelo.

```python
bow = CountVectorizer(tokenizer=tokenizer, stop_words=list(stop_words), lowercase=True)
```

```python
X_bow = bow.fit_transform(X_train)
```

```python
len(bow.vocabulary_)
```

### 3.2.2 Entrenando el modelo BoW

Estas lneas de cdigo se utilizan para entrenar un modelo de clasificacin de bosque aleatorio utilizando la representacin de bolsa de palabras X_bow y las etiquetas de sentimiento y_train. El modelo se ajusta utilizando la funcin fit del RandomForestClassifier.

```python
bow_model = RandomForestClassifier(random_state = 3)
```

```python
bow_model.fit(X_bow, y_train)
```

Luego, se utiliza la funcin feature_importances_ para obtener la importancia de cada palabra en el modelo de clasificacin de bosque aleatorio. Con la ayuda de la funcin plot.barh, se crea un grfico de barras que muestra las 20 palabras ms importantes para la clasificacin.

```python
# La 'feature importance' en modelos basados en rboles de decisin indican cuales son las caractersticas ms importantes al momento de tomar una decisin
pd.Series(bow_model.feature_importances_, index = bow.vocabulary_).sort_values().tail(20).plot.barh(figsize = (15, 10))
```

Finalmente, se imprimen algunos detalles del modelo entrenado, incluyendo el nmero de rboles utilizados y la profundidad media de los rboles, y se utiliza el modelo entrenado para hacer predicciones en los conjuntos de entrenamiento y prueba utilizando la funcin predict.

```python
bow_estimators = bow_model.estimators_
print('Number of trees:', len(bow_estimators))
print('Trees depth (mean):', np.mean([tree.get_depth() for tree in bow_estimators]))
```

```python
y_train_bow_predict = bow_model.predict(X_bow)
y_test_bow_predict = bow_model.predict(bow.transform(X_test))
```

### 3.2.3 Validacin cuantiativa de BoW

#### 3.2.3.1 Matriz de confusin 

```python
ConfusionMatrixDisplay.from_predictions(y_train, y_train_bow_predict)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_test_bow_predict)
```

Estos resultados indican que el modelo tiene una precisin perfecta, es decir, que no comete errores al predecir el sentimiento de las reseas. La mtrica de precisin se refiere a la proporcin de casos positivos que fueron identificados correctamente por el modelo, mientras que el recall se refiere a la proporcin de casos positivos totales que fueron identificados correctamente. El F1-score es una mtrica que combina la precisin y el recall en una sola medida.

```python
if len(bow_model.classes_) == 2:
    print('Precision:', precision_score(y_train, y_train_bow_predict, pos_label = 1))
    print('Recall:', recall_score(y_train, y_train_bow_predict, pos_label = 1))
    print('F1:', f1_score(y_train, y_train_bow_predict, pos_label = 1))
else:
    # Para casos no binarios, es comn calcular las mtricas para cada clase
    print('Precision:', precision_score(y_train, y_train_bow_predict, average = None))
    print('Recall:', recall_score(y_train, y_train_bow_predict, average = None))
    print('F1:', f1_score(y_train, y_train_bow_predict, average = None))
```

Estos resultados indican la precisin, recall y F1-score del modelo BoW en el conjunto de prueba. La precisin mide la fraccin de verdaderos positivos en relacin con todos los clasificados como positivos, el recall mide la fraccin de verdaderos positivos en relacin con todos los que son positivos, y el F1-score es una medida de la precisin y el recall combinados.

En este caso, los resultados indican que el modelo BoW tiene una precisin del 83.05%, un recall del 83.75% y un F1-score del 83.40% en el conjunto de prueba. Esto sugiere que el modelo tiene un buen rendimiento en la clasificacin de las reseas.

El cdigo est inicializando una instancia del mtodo HashingVectorizer con 2^16 caractersticas (n_features=2^16). El propsito del HashingVectorizer es convertir un conjunto de documentos de texto en una matriz de caractersticas numricas, utilizando una funcin de hash. Esto es til para procesar grandes cantidades de datos de texto en lnea, ya que no requiere almacenar un diccionario de caractersticas como lo hace CountVectorizer o TfidfVectorizer. En su lugar, las caractersticas se calculan sobre la marcha a medida que se procesa el texto.

El nmero de caractersticas que se deben utilizar se establece mediante el parmetro n_features. En este caso, se ha establecido en 2^16 para permitir la creacin de una matriz de caractersticas lo suficientemente grande como para capturar informacin relevante de los datos de texto.

# 5. rboles de decisin

El algoritmo de rboles de decisin es un mtodo de aprendizaje supervisado utilizado para clasificar datos en categoras o predecir valores numricos. En este algoritmo, se construye un rbol de decisiones a partir de los datos de entrenamiento, donde cada nodo del rbol representa una caracterstica del conjunto de datos y las ramas representan las posibles respuestas a esa caracterstica. El rbol se construye de forma recursiva, dividiendo el conjunto de datos en subconjuntos ms pequeos y homogneos segn ciertas caractersticas hasta alcanzar una determinada profundidad o un criterio de parada.

Este algoritmo puede resultar til para entrenar modelos de clasificacin de reseas en negativas y positivas porque permite identificar las caractersticas ms relevantes para hacer la clasificacin, como las palabras ms utilizadas en las reseas positivas o negativas. Al utilizar tcnicas de procesamiento de texto como BoW, TF-IDF y HashingVectorizer, se pueden representar las reseas en un formato que pueda ser utilizado por el algoritmo de rboles de decisin. Por ejemplo, BoW representa las reseas como vectores que indican la frecuencia de cada palabra en el conjunto de datos, mientras que TF-IDF le da ms peso a las palabras menos comunes y HashingVectorizer representa las reseas como vectores de valores binarios.

Al utilizar el algoritmo de rboles de decisin en conjunto con estas tcnicas de procesamiento de texto, se puede lograr una clasificacin ms precisa y efectiva de las reseas en negativas y positivas. Adems, el algoritmo de rboles de decisin tambin permite visualizar el proceso de toma de decisiones del modelo, lo que puede ser til para entender cmo se estn haciendo las clasificaciones y mejorar el modelo en el futuro.
